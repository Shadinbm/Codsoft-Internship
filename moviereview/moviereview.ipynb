# Model building and evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
dfm=pd.read_csv(r"C:\Users\bmsha\Documents\moviereview\model_imdb.csv")
dfm.head()
```

```python
dfm['Director'].unique()
```

```python
dfm['Genre'].unique()
```

```python
le=LabelEncoder()
dfm['Genre']=le.fit_transform(dfm['Genre'])
for label,i in enumerate(le.classes_):
    print(label,":",i)
```

> <b>Label Guide for Genre

```python
le=LabelEncoder()
dfm['Director']=le.fit_transform(dfm['Director'])
for label,i in enumerate(le.classes_):
    print(label,":",i)
```

> <b>Label guide for Directors

### Model Validation

```python
fet=['Duration','Genre','Votes','Director']
x=dfm[fet]
y=dfm['Rating']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
xtrain
```

```python
ytrain
```

# Linear Regression Model

```python
lm=LinearRegression()
lm.fit(xtrain,ytrain)
ypred=lm.predict(xtest)
```

```python
print("MAE",mean_absolute_error(ytest,ypred))
print("MSE",mean_squared_error(ytest,ypred))
print("r2Score",r2_score(ytest,ypred))
```

```python
lm.predict(xtest.iloc[[1]])
```

```python
ytest.iloc[1]
```

> <b> linear regression model shows weak predictive ability, with an MAE of 0.59 and an MSE of 0.94, indicating that predictions deviate from actual values by around half a unit on average with some larger errors. The very low R² score (0.028) highlights that the model explains less than 3% of the variance in the target variable, meaning it is only slightly better than using the mean as a prediction

# RandomForest Regressor

```python
rm=RandomForestRegressor(n_estimators=100,random_state=42)
rm.fit(xtrain,ytrain)
ypredr=rm.predict(xtest)
```

```python
print("MAE",mean_absolute_error(ytest,ypredr))
print("MSE",mean_squared_error(ytest,ypredr))
print("r2Score",r2_score(ytest,ypredr))
```

```python
rm.predict(xtest.iloc[[1]])
```

```python
ytest.iloc[1]
```

> <b>Random forestregression results show some improvement compared to the previous run. The MAE of 0.51 indicates that predictions are, on average, about half a unit away from actual values, while the MSE of 0.84 reflects slightly smaller squared errors. Most importantly, the R² score has increased to 0.129, meaning the model now explains about 13% of the variance in the target variable—still low, but notably better than before. This suggests the model is learning some useful patterns,

# SVR 

```python
svr=SVR(kernel='rbf',degree=3,gamma='scale', C=100, epsilon=0.1)
svr.fit(xtrain,ytrain)
ypreds=svr.predict(xtest)
```

```python
print("MAE",mean_absolute_error(ytest,ypreds))
print("MSE",mean_squared_error(ytest,ypreds))
print("r2Score",r2_score(ytest,ypreds))
```

```python
svr.predict(xtest.iloc[[1]])
```

```python
ytest.iloc[1]
```

> <b> The model’s performance is weak, with an MAE of 0.58 indicating that predictions deviate from actual values by more than half a unit on average, and an MSE of 0.95 showing the presence of relatively large squared errors. The very low R² score of 0.015 reveals that the model explains only about 1.5% of the variance in the target variable, suggesting that it fails to capture meaningful patterns. This highlights the need for further improvements through better feature engineering, parameter tuning, or exploring alternative modeling approaches

# DecisionTreeRegressor

```python
dr=DecisionTreeRegressor(criterion='squared_error',
    max_depth=5,
    min_samples_split=10,
    min_samples_leaf=4,
    max_features='sqrt',
    random_state=42)
dr.fit(xtrain,ytrain)
ypredd=dr.predict(xtest)
```

```python
print("MAE",mean_absolute_error(ytest,ypredd))
print("MSE",mean_squared_error(ytest,ypredd))
print("r2Score",r2_score(ytest,ypredd))
```

```python
dr.predict(xtest.iloc[[1]])
```

```python
ytest.iloc[1]
```

> <b>The model shows slight improvement compared to earlier results. The MAE of 0.57 indicates that predictions deviate from actual values by a little more than half a unit on average, while the MSE of 0.87 reflects somewhat lower squared errors. The R² score of 0.099 means the model explains about 10% of the variance in the target variable—better than before but still weak overall.

## Conclusion

<h3 style='background-color:cyan'>Among the models tested, linear regression provided very weak predictive power with only a small portion of variance explained. Support Vector Regression performed the worst, with almost no ability to capture meaningful patterns. A single decision tree showed slight improvement but still remained limited due to its tendency to overfit and poor generalization. In contrast, ensemble methods like random forests are generally the most effective choice, as they combine multiple trees to capture nonlinear relationships and usually deliver stronger and more reliable predictive performance compared to the other approaches.</h3>
